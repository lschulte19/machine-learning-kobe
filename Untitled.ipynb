```python
import pandas as pd
from pathlib import Path

# Carregar os dados
path = Path.home() / "machine-learning-kobe" / "data" / "raw" / "dataset_kobe_dev.parquet"
df = pd.read_parquet(path)

# 1. Tipos de dados básicos
print("Tipos de dados:\n", df.dtypes)

# 2. Estatísticas descritivas
print("\nEstatísticas descritivas:\n", df.describe(include='all'))

# 3. Análise detalhada por coluna
for col in df.columns:
    print(f"\n--- Análise da coluna '{col}' ---")
    print("Tipo:", df[col].dtype)
    print("Valores únicos:", df[col].nunique())
    print("Valores nulos:", df[col].isnull().sum())
    
    # Para colunas categóricas/strings
    if df[col].dtype == 'object':
        print("\nAmostra de valores:", df[col].unique()[:10])
    
    # Para colunas numéricas
    elif pd.api.types.is_numeric_dtype(df[col]):
        print("Média:", df[col].mean())
        print("Mediana:", df[col].median())
```
